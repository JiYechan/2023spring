```python
#라이브러리 삽입
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb

#파일 불러오기
iris=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',header=None)

#칼럼별로 이름 붙이기
iris.columns= ['sl','sw','pl','pw','name']

#꽃 종류와 개수 확인
print(iris['name'].value_counts()) 


# step_function 만들기
def step_func (threshold, x):
    if x<threshold:
        return 0
    return 1

#number of trial 정하기
n_trials=1000 

# 테스트할 수
n_test= 20
# learning rate 정하기
alpha=0.01

# 100개의 데이터(2가지 종류의 꽃)로 잘라내기
temp=iris.copy().truncate(after=99) 
```

```python
#1 입력값 sl과 sw의 비교

X1=iris.loc[0:99,'sl'] 
X2=iris.loc[0:99,'sw']
X=pd.concat([X1,X2],axis=1)

#입력값 넘파이 어레이로 변환 
X=X.to_numpy()   

#100개의 타켓 만들기(Setosa)
T=iris.loc[0:99,'name'] 
T=T.to_numpy()  
T=np.where(T=='Iris-setosa',1,0)


# 정규 분포로 가중치 정하기
W=np.random.randn(3)




##학습 데이터와 테스트 데이터 분리하기

#데이터 섞고 인덱스 나누기
indices = np.random.permutation(len(X))

train_indices = indices[:80]
test_indices = indices[80:]

# 학습 데이터와 테스트 데이터 분리
X_train, T_train = X[train_indices], T[train_indices]
X_test, T_test = X[test_indices], T[test_indices]

# 학습시키기
hit_array=[]  

for i in range(n_trials):
    for j in range(len(X_train)):
        s=np.dot(X_train[j],W[1:])+W[0]  
        delta=alpha*(T_train[j]-step_func(0,s)) 
        W[1:]+=delta*X_train[j]  
        W[0]+=delta
        
        
# hit rate 계산
    hit=0
    for i in range(n_test):
        
        s=np.dot(X_test[i],W[1:])+W[0]
        delta=(T_test[i]-step_func(s,0)) 
        if delta!=0:
            hit+=1
    
    hit_array.append(hit)


hit_rate = np.array(hit_array) / n_test
plt.plot(hit_rate)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np



# 퍼셉트론 클래스 만들기
class mPerceptron:
    
    #input, hidde, output, learning rate
    def __init__(self, i, h, o, a):
        self.i = i
        self.h = h
        self.o = o
        self.a = a
        
        # 가중치 랜덤 
        self.Wih=np.random.rand(self.i,self.h) 
        self.Who=np.random.rand(self.h, self.o)
        
        #step function을 시그모이드함수로.
    def step_func(self, x):
        return 1/(1+np.exp(-x))   
        
        #학습과 훈련
    def train(self, inputs, target):
        #2차원 매트릭스로
        I=np.array(inputs, ndmin=2)
        T=np.array(target, ndmin=2)
                
        #인풋과 가중치 dot product하고 step function에 넣어줌
        hsum=np.dot(I, self.Wih)
        H=self.step_func(hsum)
        
        #그것을 두번째 웨잇메트릭스와 dot product. 
        osum=np.dot(H,self.Who)
        
        #그걸 또 step function하면 최종 아웃풋
        Y=self.step_func(osum)
        
        #이제 아웃풋과 타겟하고 비교를 해서 back propagation
        
            
        #에러 계산과  backpropagation으로 수정
        delta=T-Y
        hdelta=np.dot(delta, self.Who.T)

        self.Who += self.a*np.dot(H.T, delta*Y*(1-Y))        
        self.Wih += self.a*np.dot(I.T, hdelta*H*(1-H))
        
        # 테스트
    def test(self, inputs):
        #2차원 어레이로
        I=np.array(inputs, ndmin=2)
        
        #Forward propagation
        hsum=np.dot(I, self.Wih)
        H=self.step_func(hsum)
        osum=np.dot(H,self.Who)
        Y=self.step_func(osum)
        
        return Y
        
```

```python
#파일 불러오기
iris=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',header=None)

#이름 변경
temp=pd.get_dummies(iris[4])  

temp.replace({1:0.99,0:0.01},inplace=True)

#np로 변경
X=iris.iloc[0:150,:4].to_numpy()
X/=[7.9,4.4,6.9,2.5]
#최댓값으로 나눠주기


#타켓 np로 변경
T=temp.iloc[0:150,:3].to_numpy()



```

```python
#parameter 정하기
n_trials=300 
n_test=20  
a=0.02 

i=4  
h=10
o=3
hit_array=[]
  

my=mPerceptron(i,h,o,a)


#퍼셉트론 훈련
for j in range(n_trials):   
    for i in range (len(X)):   
            my.train(X[i],T[i]) 
    
    #퍼셉트론 테스트
    hit=0
    for i in range (n_test):   
        s=np.random.randint(0,149)      
        if((np.argmax(T[s])) == (np.argmax(my.test(X[s])))):
            hit+=1 
    
    hit_array.append(hit)

plt.plot(hit_array)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np



# 퍼셉트론 클래스 만들기
class mPerceptron:
    
    #input, hidde, output, learning rate
    def __init__(self, i, h, o, a):
        self.i = i
        self.h = h
        self.o = o
        self.a = a
        
        # 가중치 랜덤 
        self.Wih=np.random.rand(self.i,self.h) 
        self.Who=np.random.rand(self.h, self.o)
        
    #step function을 시그모이드함수로.
    def step_func(self, x):
        return 1/(1+np.exp(-x))   
    
        #학습과 훈련
    def train(self, inputs, target):

        I=np.array(inputs, ndmin=2)
        T=np.array(target, ndmin=2)
                
        #인풋과 가중치 dot product하고 step function에 넣어줌
        hsum=np.dot(I, self.Wih)
        H=self.step_func(hsum)
        
        #그것을 두번째 웨잇메트릭스와 dot product. 
        osum=np.dot(H,self.Who)
        
        #그걸 또 step function하면 최종 아웃풋
        Y=self.step_func(osum)
        
        #이제 아웃풋과 타겟하고 비교를 해서 back propagation
        
            
        #에러 계산과  backpropagation으로 수정
        delta=T-Y
        hdelta=np.dot(delta, self.Who.T)

        self.Who += self.a*np.dot(H.T, delta*Y*(1-Y))        
        self.Wih += self.a*np.dot(I.T, hdelta*H*(1-H))
        
        # 테스트
    def test(self, inputs):
        #2차원 어레이로
        I=np.array(inputs, ndmin=2)
        
        #Forward propagation
        hsum=np.dot(I, self.Wih)
        H=self.step_func(hsum)
        osum=np.dot(H,self.Who)
        Y=self.step_func(osum)
        
        return Y
        
```
