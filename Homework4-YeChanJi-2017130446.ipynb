```python
#Question 1 그리디와 엡실론 그리디의 비교

import numpy as np
import matplotlib.pyplot as plt

#패러미터 설정
n_arms=10
n_steps= 1000
n_trials=2000


#우선 엡실론 0 설정(greedy policy)
epsilon=0

#Result가 높은 reward를 얻었는지
Result_array=np.zeros(n_steps)  

#Result가 최적의 선택을 했는지
Result_array2=np.zeros(n_steps)   

#표준 정규분포에 따라, average probability of reward 정함
P_arm=np.random.randn(n_arms)    



#트라이얼을 제일 바깥에서 돌려줌
for i in range(n_trials):
    # t시점에서의 value값
    Q_arm=np.zeros(n_arms)     
    # 각 arm의 시행횟수
    N_visits=np.zeros(n_arms)   
    
    
    #step을 돌며 reward 얻는 과정
    for j in range(n_steps):
        
        #평균값을 P_arm으로, 표준편차 1로 하는 정규분포값으로 R_arm 설정
        R_arm=np.random.normal(P_arm,1,n_arms)  
        
        #엡실론 그리디 policy
        if np.random.rand(1) < epsilon:        
            action=np.random.choice(n_arms,1)
        
        #그리디 policy
        #똑같은 값이 있으면, 인덱스 랜덤 선택
        else:
            max_indices = np.where(Q_arm == np.max(Q_arm))[0]
            action = np.random.choice(max_indices)
            # 수정 전, action=np.argmax(Q_arm) 
        
        #처음은 바로 R_arm을 Q_arm에 넣어줌     
        if N_visits[action]==0:                 
            Q_arm[action]+=R_arm[action]   
        #이후부터는 incremental한 공식 사용
        else:
            Q_arm[action]+=(R_arm[action]-Q_arm[action])/N_visits[action]   
        
        #방문횟수 1 늘려줌
        N_visits[action]+=1
        
        #Result_array에 reward값에 더해줌
        Result_array[j]+=R_arm[action]
        
        #Result_array2에 optimal한 선택을 했는지 더해줌
        if np.argmax(R_arm)==action:
            Result_array2[j]+=1
    
#plt.plot(Result_array/n_trials, color='green')
plt.plot(Result_array2/n_trials, color='green')

#---------------------------------------------------------------------------------------------

#엡실론 수정해서 똑같은 코드 그대로 사용
epsilon=0.1

#Result가 높은 reward를 얻었는지
Result_array=np.zeros(n_steps)  

#Result가 최적의 선택을 했는지
Result_array2=np.zeros(n_steps)   

#표준 정규분포에 따라, average probability of reward 정함
P_arm=np.random.randn(n_arms)    



#트라이얼을 제일 바깥에서 돌려줌
for i in range(n_trials):
    # t시점에서의 value값
    Q_arm=np.zeros(n_arms)     
    # 각 arm의 시행횟수
    N_visits=np.zeros(n_arms)   
    
    
    #step을 돌며 reward 얻는 과정
    for j in range(n_steps):
        
        #평균값을 P_arm으로, 표준편차 1로 하는 정규분포값으로 R_arm 설정
        R_arm=np.random.normal(P_arm,1,n_arms)  
        
        #엡실론 그리디 policy
        if np.random.rand(1) < epsilon:        
            action=np.random.choice(n_arms,1)
        
        #그리디 policy
        #똑같은 값이 있으면, 인덱스 랜덤 선택
        else:
            max_indices = np.where(Q_arm == np.max(Q_arm))[0]
            action = np.random.choice(max_indices)
            # 수정 전, action=np.argmax(Q_arm) 
        
        #처음은 바로 R_arm을 Q_arm에 넣어줌     
        if N_visits[action]==0:                 
            Q_arm[action]+=R_arm[action]   
        #이후부터는 incremental한 공식 사용
        else:
            Q_arm[action]+=(R_arm[action]-Q_arm[action])/N_visits[action]   
        
        #방문횟수 1 늘려줌
        N_visits[action]+=1
        
        #Result_array에 reward값에 더해줌
        Result_array[j]+=R_arm[action]
        
        #Result_array2에 optimal한 선택을 했는지 더해줌
        if np.argmax(R_arm)==action:
            Result_array2[j]+=1

#plt.plot(Result_array/n_trials, color='red')
plt.plot(Result_array2/n_trials, color='blue')


#---------------------------------------------------------------------------------------------

#엡실론 값만  수정해서 그대로 사용.
epsilon=0.01

#Result가 높은 reward를 얻었는지
Result_array=np.zeros(n_steps)  

#Result가 최적의 선택을 했는지
Result_array2=np.zeros(n_steps)   

#표준 정규분포에 따라, average probability of reward 정함
P_arm=np.random.randn(n_arms)    



#트라이얼을 제일 바깥에서 돌려줌
for i in range(n_trials):
    # t시점에서의 value값
    Q_arm=np.zeros(n_arms)     
    # 각 arm의 시행횟수
    N_visits=np.zeros(n_arms)   
    
    
    #step을 돌며 reward 얻는 과정
    for j in range(n_steps):
        
        #평균값을 P_arm으로, 표준편차 1로 하는 정규분포값으로 R_arm 설정
        R_arm=np.random.normal(P_arm,1,n_arms)  
        
        #엡실론 그리디 policy
        if np.random.rand(1) < epsilon:        
            action=np.random.choice(n_arms,1)
        
        #그리디 policy
        #똑같은 값이 있으면, 인덱스 랜덤 선택
        else:
            max_indices = np.where(Q_arm == np.max(Q_arm))[0]
            action = np.random.choice(max_indices)
            # 수정 전, action=np.argmax(Q_arm) 
        
        #처음은 바로 R_arm을 Q_arm에 넣어줌     
        if N_visits[action]==0:                 
            Q_arm[action]+=R_arm[action]   
        #이후부터는 incremental한 공식 사용
        else:
            Q_arm[action]+=(R_arm[action]-Q_arm[action])/N_visits[action]   
        
        #방문횟수 1 늘려줌
        N_visits[action]+=1
        
        #Result_array에 reward값에 더해줌
        Result_array[j]+=R_arm[action]
        
        #Result_array2에 optimal한 선택을 했는지 더해줌
        if np.argmax(R_arm)==action:
            Result_array2[j]+=1
    
#plt.plot(Result_array/n_trials, color='red')
plt.plot(Result_array2/n_trials, color='red')



#---------------------------------------------------------------------------------------------

#엡실론 값만  수정해서 그대로 사용.
epsilon=0.5

#Result가 높은 reward를 얻었는지
Result_array=np.zeros(n_steps)  

#Result가 최적의 선택을 했는지
Result_array2=np.zeros(n_steps)   

#표준 정규분포에 따라, average probability of reward 정함
P_arm=np.random.randn(n_arms)    



#트라이얼을 제일 바깥에서 돌려줌
for i in range(n_trials):
    # t시점에서의 value값
    Q_arm=np.zeros(n_arms)     
    # 각 arm의 시행횟수
    N_visits=np.zeros(n_arms)   
    
    
    #step을 돌며 reward 얻는 과정
    for j in range(n_steps):
        
        #평균값을 P_arm으로, 표준편차 1로 하는 정규분포값으로 R_arm 설정
        R_arm=np.random.normal(P_arm,1,n_arms)  
        
        #엡실론 그리디 policy
        if np.random.rand(1) < epsilon:        
            action=np.random.choice(n_arms,1)
        
        #그리디 policy
        #똑같은 값이 있으면, 인덱스 랜덤 선택
        else:
            max_indices = np.where(Q_arm == np.max(Q_arm))[0]
            action = np.random.choice(max_indices)
            # 수정 전, action=np.argmax(Q_arm) 
        
        #처음은 바로 R_arm을 Q_arm에 넣어줌     
        if N_visits[action]==0:                 
            Q_arm[action]+=R_arm[action]   
        #이후부터는 incremental한 공식 사용
        else:
            Q_arm[action]+=(R_arm[action]-Q_arm[action])/N_visits[action]   
        
        #방문횟수 1 늘려줌
        N_visits[action]+=1
        
        #Result_array에 reward값에 더해줌
        Result_array[j]+=R_arm[action]
        
        #Result_array2에 optimal한 선택을 했는지 더해줌
        if np.argmax(R_arm)==action:
            Result_array2[j]+=1
    
#plt.plot(Result_array/n_trials, color='black')
plt.plot(Result_array2/n_trials, color='black')

```
